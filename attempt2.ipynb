Create a diagram where the x-axis is the date and the y-axis is for multiple lines Open, High, Low, Close, and Volume

```python
import matplotlib.dates as mdates

# convert date to matplotlib dates
btc_df['Date'] = pd.to_datetime(btc_df['Date'])
btc_df['Date'] = btc_df['Date'].apply(mdates.date2num)

```

```python
def candlestick_chart(last_days: int, btc_df: pd.DataFrame, predict: bool = False):
    btc_df = btc_df.iloc[-last_days:]


    # create a figure and axis object
    fig, ax = plt.subplots()

    # plot the candlesticks
    ax.plot(btc_df['Date'], btc_df['Open'], color='green', linewidth=1, label='Open')
    ax.plot(btc_df['Date'], btc_df['Close'], color='black', linewidth=1, label='Close')
    ax.vlines(btc_df['Date'], btc_df['Low'], btc_df['High'], color='blue', linewidth=1)
    
    if predict:
        ax.plot(btc_df['Date'], btc_df['Predict Open'], color='red', linewidth=1, label='Predict Open')
        ax.plot(btc_df['Date'], btc_df['Predict Close'], color='orange', linewidth=1, label='Predict Close')
        ax.vlines(btc_df['Date'], btc_df['Predict Low'], btc_df['Predict High'], color='purple', linewidth=1)

    # format the x-axis as dates
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

    # set the axis labels and title
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.set_title('Candlestick Chart')

    # add a legend
    ax.legend()

    # show the chart
    plt.show()
    
def volume(data = pd.DataFrame, last_days: int = 2, predict: bool = False):
    data = data.iloc[-last_days:]
    if predict:
        sns.barplot(x='Date', y='Predict Volume', data=data)
        sns.barplot(x='Date', y='Volume', data=data)
    else:
        sns.barplot(x='Date', y='Volume', data=data)
    

candlestick_chart(5, btc_df)
volume(btc_df, 5)

```

```python
btc_df['Predict Open'] = regressor.predict(X)
```

I think it would be fair to say the starting point of predictions should be the opening price since I'll only be able to collect this data every 24hours. Use Multiple Linear or Multi-Polynomial Regression models for determining the 'Open' price based on the day before's 'Close', 'High', 'Low', and 'Volume' features. 

Once, the next day's 'Open' price has been predicted I think it's safe to assume a simple Linear or Poylnomial Regression model could be used to predict the 'Close' price based on the predicted 'Open' price. 

Next, we can cover 2 features with 2 features. For now, I'm going to use Multiple Linear Regression Models to predict the 'High' price based on the fields 'Open' and 'Close'. Then, we will use the same model for predicting the 'Low' price based on the 'Open' and 'Close' price.

Lastly, we can now predict 'Volume' based on 'High', 'Low', 'Open', and 'Close'. 

```python
# Open Model --> Predict `Open` prce based on `High`, `Low`, `Close`, and `Volume`
open_regressor = LinearRegression(fit_intercept=True)
close_regressor = LinearRegression(fit_intercept=True)
high_regressor = LinearRegression(fit_intercept=True)
low_regressor = LinearRegression(fit_intercept=True)
volume_regressor = LinearRegression(fit_intercept=True)

OpenX = btc_df[['High', 'Low', 'Close', 'Volume']]
openy = btc_df['Open']

OpenX_train, OpenX_test, openy_train, openy_test = train_test_split(OpenX, openy, test_size=0.2, random_state=42)

open_regressor.fit(OpenX_train, openy_train)

open_y_pred = open_regressor.predict(OpenX_test)

open_score = open_regressor.score(OpenX_test, openy_test)
open_score
```

```python
CloseX = btc_df[['Open']]
closey = btc_df['Close']

CloseX_train, CloseX_test, closey_train, closey_test = train_test_split(CloseX, closey, test_size=0.2, random_state=42)

close_regressor.fit(CloseX_train, closey_train)

close_y_pred = close_regressor.predict(CloseX_test)

close_score = close_regressor.score(CloseX_test, closey_test)
close_score
```

```python
HighX = btc_df[['Open', 'Close']]
highy = btc_df['High']

HighX_train, HighX_test, highy_train, highy_test = train_test_split(HighX, highy, test_size=0.2, random_state=42)

high_regressor.fit(HighX_train, highy_train)

high_y_pred = high_regressor.predict(HighX_test)

high_score = high_regressor.score(HighX_test, highy_test)
high_score
```

```python
LowX = btc_df[['Open', 'Close']]
lowy = btc_df['Low']

LowX_train, LowX_test, lowy_train, lowy_test = train_test_split(LowX, lowy, test_size=0.2, random_state=42)

low_regressor.fit(LowX_train, lowy_train)

low_y_pred = low_regressor.predict(LowX_test)

low_score = low_regressor.score(LowX_test, lowy_test)
low_score
```

```python
VolumeX = btc_df[['Open', 'High', 'Low', 'Close']]
volumey = btc_df['Volume']

VolumeX_train, VolumeX_test, volumey_train, volumey_test = train_test_split(VolumeX, volumey, test_size=0.2, random_state=42)

volume_regressor.fit(VolumeX_train, volumey_train)

volume_y_pred = volume_regressor.predict(VolumeX_test)

volume_score = volume_regressor.score(VolumeX_test, volumey_test)
volume_score
```
